```python
import numpy as np
import pandas as pd
import matplotlib as mpl
dr=pd.read_csv(r"scrap-jurica-lamudi.csv",dtype="string")
dr['precio'] = dr['precio'].fillna('0')
dr['precio'] = dr['precio'].apply(lambda x: x.replace('$', '').replace(',', '').replace('MN','').replace(' ','').replace('NM','')).astype(float)
dr['precio'] = dr['precio'].replace(0,np.NaN)
print("El promedio de renta de una casa en jurica es de: $",round(dr['precio'].mean(),3))
print("la desviacion estandar de los precios de venta de una casa en jurica es de: $",round(dr['precio'].std(),3))
print(dr["precio"].plot(style = "o",ms = 1,xlabel="casas",ylabel = "precio"))
```
