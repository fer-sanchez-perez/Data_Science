```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as mpl
dr=pd.read_csv(r"departamentos.csv",dtype="string")

dr['precio'] = dr['precio'].fillna('0')
dr['precio'] = dr['precio'].apply(lambda x: x.replace('$', '').replace(',', '').replace('MN','').replace(' ','').replace('NM','')).astype(float)
dr['precio'] = dr['precio'].replace(0,np.NaN)

print("El promedio de renta de un departamento es de: $",round(dr['precio'].mean(),3))
print("la desviacion estandar de los precios de renta de un departamento es de: $",round(dr['precio'].std(),3))
print("El precio mas alto en departamentos es de: $",dr['precio'].max())
print("El precio mas bajo en departamentos es de: $",dr['precio'].min())
print("La mediana en departamentos es de: $",dr['precio'].median())
print("La moda en departamentos es de: $",dr['precio'].mode()[0])


dr['recamaras'] = dr['recamaras'].fillna('0')
dr['recamaras'] = dr['recamaras'].apply(lambda x: x.replace(' ','')).astype(int)
dr['recamaras'] = dr['recamaras'].replace(0,np.NaN)

print("El promedio de recamaras de un departamento es de: $",round(dr['recamaras'].mean(),1))

dr['baños'] = dr['baños'].fillna('0')
dr['baños'] = dr['baños'].apply(lambda x: x.replace(' ','')).astype(int)
dr['baños'] = dr['baños'].replace(0,np.NaN)

print("El promedio de baños de un departamento es de: $",round(dr['baños'].mean(),1))

dr['mconstruidos'] = dr['mconstruidos'].fillna('0')
dr['mconstruidos'] = dr['mconstruidos'].apply(lambda x: x.replace(' ','')).astype(int)
dr['mconstruidos'] = dr['mconstruidos'].replace(0,np.NaN)

print("El promedio de m² construidos de un departamento es de: $",round(dr['mconstruidos'].mean(),3))

colonias=['El Refugio', 'Juriquilla,', 'Milenio', 'El Mirador','El campanario', 'Jurica', 'Centro Sur', 'Cumbres del lago', 'Candiles', 'El Pueblico']
promdepas=[dr.loc[dr['web-scraper-start-url'].str.contains("refugio"),'precio'].mean(),
            dr.loc[dr['web-scraper-start-url'].str.contains("juriquilla"),'precio'].mean(),
            dr.loc[dr['web-scraper-start-url'].str.contains("milenio"),'precio'].mean(),
            dr.loc[dr['web-scraper-start-url'].str.contains("mirador"),'precio'].mean(),
            dr.loc[dr['web-scraper-start-url'].str.contains("campanario"),'precio'].mean(),
            dr.loc[dr['web-scraper-start-url'].str.contains("jurica"),'precio'].mean(),
            dr.loc[dr['web-scraper-start-url'].str.contains("sur"),'precio'].mean(),
            dr.loc[dr['web-scraper-start-url'].str.contains("cumbres"),'precio'].mean(),
            dr.loc[dr['web-scraper-start-url'].str.contains("candiles"),'precio'].mean(),
            dr.loc[dr['web-scraper-start-url'].str.contains("pueblito"),'precio'].mean()]
preciosprom = mpl.figure()
prom = mpl.figure()
prom.set_figheight(10)
prom.set_figwidth(20)
mpl.title('Promedio de precios por colonia', fontdict={'size': 16})
mpl.bar(colonias,promdepas,color="#c237db")

pre = mpl.figure()
pre.set_figheight(10)
pre.set_figwidth(20)
mpl.plot(dr['precio'].sort_values(),'b^')
mpl.annotate('Mediana:18000',xy=(800,18000),xytext=(1270,40000),arrowprops=dict())
mpl.annotate('Moda: 18000',xy=(800,18000),xytext=(270,40000),arrowprops=dict())

print('\n\nTop 5 de departamentos con mayor valor')
print(dr[['titulo','precio']].sort_values(by='precio',ascending=False).head())

dr['precioxmetro'] = dr['precio']/dr['mconstruidos']
print("\nEl precio de renta por m² en El Refugio es de $",round(dr.loc[dr['web-scraper-start-url'].str.contains("refugio"),'precioxmetro'].mean(),3))
print("El precio de renta por m² en Juriquilla es de $",round(dr.loc[dr['web-scraper-start-url'].str.contains("juriquilla"),'precioxmetro'].mean(),3))
print("El precio de renta por m² en Milenio III es de $",round(dr.loc[dr['web-scraper-start-url'].str.contains("milenio"),'precioxmetro'].mean(),3))
print("El precio de renta por m² en El Mirador es de $",round(dr.loc[dr['web-scraper-start-url'].str.contains("mirador"),'precioxmetro'].mean(),3))
print("El precio de renta por m² en El Campanario es de $",round(dr.loc[dr['web-scraper-start-url'].str.contains("campanario"),'precioxmetro'].mean(),3))
print("El precio de renta por m² en Jurica es de $",round(dr.loc[dr['web-scraper-start-url'].str.contains("jurica"),'precioxmetro'].mean(),3))
print("El precio de renta por m² en Centro Sur es de $",round(dr.loc[dr['web-scraper-start-url'].str.contains("sur"),'precioxmetro'].mean(),3))
print("El preci de renta por m² en Cumbre del Lago es de $",round(dr.loc[dr['web-scraper-start-url'].str.contains("cumbres"),'precioxmetro'].mean(),3))
print("El precio de renta por m² en Candiles es de $",round(dr.loc[dr['web-scraper-start-url'].str.contains("candiles"),'precioxmetro'].mean(),3))
```
